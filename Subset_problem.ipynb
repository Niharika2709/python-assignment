#*Subset Selection Problem*


1. Problem Statement:
Find all the subsets from a set of numbers whose sum is zero.

    ***Constraint: Subset size must be 5***

Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
import random as r


Set = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

SetSize = 5
Iterations = 100000
ResultList = set()

for i in range(Iterations):
    Chromosome = r.sample(Set, SetSize)
    if sum(Chromosome) == 0:
        ResultList.add(tuple(sorted(Chromosome)))


print(f"\nTotal unique zero-sum subsets of size {SetSize}: {len(ResultList)}\n")
for subset in sorted(ResultList):
    print(subset)


print("\nTotal Sets Found:", len(ResultList))
```

#*2. Problem Statement:*
Find all the subsets from a set of numbers whose sum is zero.

   ***Constraint: Subset size must be 3 to 6 only***

Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

```python
from itertools import combinations

numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

unique_combinations = {}

for size in range(3, 7):
    for subset in combinations(numbers, size):
        if sum(subset) == 0:
            sorted_subset = tuple(sorted(subset))
            unique_combinations[sorted_subset] = True


print(f"\n Total zero-sum subsets (size 3 to 6): {len(unique_combinations)}\n")
for subset in sorted(unique_combinations):
    print(subset)
```
