1. Convert 200MB, 400 MB, 600 MB, 800 MB, and 1000MB text files to upper case.

 Step 1: Prepare Test Files

```python
def generate_file(filename, target_mb):
    line = "This is a sample line of text to convert to uppercase.\n"
    bytes_written = 0
    target_bytes = target_mb * 1024 * 1024  # MB to Bytes

    with open(filename, "w", encoding="utf-8") as f:
        while bytes_written < target_bytes:
            f.write(line)
            bytes_written += len(line.encode("utf-8"))

sizes = [200, 400, 600, 800, 1000]
for size in sizes:
    generate_file(f"input_{size}MB.txt", size)
```

Step 2: Uppercase Conversion in Each Language

```python
import time as t

def uppercase(input_file, output_file):
    start = t.time()

    with open(input_file, "r", encoding="utf-8") as infile:
        text = infile.read()

    with open(output_file, "w", encoding="utf-8") as outfile:
        outfile.write(text.upper())

    end = t.time()
    return round(end - start, 2)

# Process each file
sizes = [200, 400, 600, 800, 1000]
for size in sizes:
    input_file = f"input_{size}MB.txt"
    output_file = f"output_{size}MB.txt"
    taken = uppercase(input_file, output_file)
    print(f"{size}MB file converted in {taken} seconds.")
```
